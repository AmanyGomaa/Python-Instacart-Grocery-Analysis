#### creating a flag for customers with many orders. to fing the loyal customers

Split the data into groups based on the “user_id” column.
Apply the transform() function on the “order_number” column to generate the maximum orders for each user.
Create a new column, “max_order,” into which you’ll place the results of your aggregation.


```python
df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df.head(15)
```

#### step 3 Deriving Columns with loc()


```python
# create a flag that assigns a “loyalty” label to a user ID based on its corresponding max order value
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

### question 7

```python
# Creating a column for frequet users
ords_prods_merged_updated['Customer_frequency'] = ords_prods_merged_updated.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Creating flag on order frequency
ords_prods_merged_updated.loc[ords_prods_merged_updated['Customer_frequency'] > 20, 'frequency_flag'] = 'Non_frequent_customer'
```

```python
ords_prods_merged_updated.loc[(ords_prods_merged_updated['Customer_frequency']>10) & (ords_prods_merged_updated['Customer_frequency']<=20),'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merged_updated.loc[ords_prods_merged_updated['Customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```
